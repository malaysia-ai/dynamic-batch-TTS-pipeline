```python
import requests

headers = {
    'accept': 'application/json',
}

params = {
    "text": "Menteri Besar, Datuk Onn Hafiz Ghazi berkata, titah pertukaran cuti hujung minggu itu merupakan keputusan yang terbaik untuk rakyat dan negeri ini secara keseluruhan.",
    "file_response": "false",
    "response_format": "mp3"
}

files = {
    'audio_input': ('audio.mp3', open('../stress-test/husein-greeting.mp3', 'rb'), 'audio/mpeg'),
}

response = requests.post('http://localhost:7088/tts', headers=headers, files=files, params=params)
```

```python
import base64
import io
import soundfile as sf
import copy
```

```python
r = response.json()
r.keys(), r['stats']
```

```python
audio_binary = base64.b64decode(r['audio'])
buffer = io.BytesIO(audio_binary)
audio_array, sample_rate = sf.read(copy.copy(buffer))
len(audio_array) / sample_rate, sample_rate
```

```python
import IPython.display as ipd

ipd.Audio(audio_array, rate = sample_rate)
```

```python
sf.write('voice-cloning-husein.mp3', audio_array, sample_rate)
```
